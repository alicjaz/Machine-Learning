```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
ds_vlagun = pd.read_csv(r'C:\Users\Ala\Downloads\VLagun_Phys_Years3.csv')
```

```python
import seaborn as sns
from sklearn.linear_model import LogisticRegression

X=ds_vlagun['PSU'].values.reshape(-1,1)
y=ds_vlagun['Years'].values.reshape(-1,1)

LogR = LogisticRegression()
LogR.fit(X,np.ravel(y.astype(int)))
sns.regplot(x='PSU', y='Years', data=ds_vlagun, logistic=True)
plt.scatter(X,y)
plt.xlabel('PSU')
plt.ylabel('Years')
plt.scatter(X, LogR.predict_proba(X)[:,1])
```
