```python
plt.figure(figsize=(15,10))
correlations = data.corr()
sns.heatmap(round(correlations,2), cmap="RdBu", annot=True,  annot_kws={"size":7}, vmin=-1, vmax=1)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from scipy.spatial.distance import squareform

plt.figure(figsize=(12,5))
dissimilarity = 1 - abs(correlations)
Z = linkage(squareform(dissimilarity), 'complete')

dendrogram(Z, labels=data.columns, orientation='top',leaf_rotation=90, leaf_font_size=8)
```

```python
treshold = 0.99
labels = fcluster(Z, treshold, criterion='distance')
labels
```

```python
import numpy as np 
labels_order = np.argsort(labels)
for idx, i in enumerate(data.columns[labels_order]):
    if idx ==0:
        clustered = pd.DataFrame(data[i])
    else:
        df_to_append = pd.DataFrame(data[i])
        clustered = pd.concat([clustered, df_to_append], axis=1)
```

```python
plt.figure(figsize=(15,10))
correlations = clustered.corr()
sns.heatmap(round(correlations,2), cmap="RdBu", annot=True,  annot_kws={"size":7}, vmin=-1, vmax=1)
plt.title("Treshold: {}".format(treshold))
```

```python
plt.figure(figsize=(15,10))
for idx, t in enumerate(np.arange(0.2,1.1,0.1)): 
    plt.subplot(3,3,idx+1)
    labels = fcluster(Z, t, criterion='distance')
    labels_order = np.argsort(labels)
    for idx, i in enumerate(clustered.columns[labels_order]):
        if idx==0: 
            clustered = pd.DataFrame(data[i])
        else:
            df_to_append = pd.DataFrame(data[i])
            clustered = pd.concat([clustered, df_to_append], axis=1)

    correlations = clustered.corr()
    sns.heatmap(round(correlations,2), cmap="RdBu", annot=True,  annot_kws={"size":7}, vmin=-1, vmax=1)
    plt.title("Treshold: {}".format(t))
```

```python
sns.clustermap(correlations, method="complete", cmap='RdBu', vmin=-1, vmax=1, annot=True, annot_kws={"size":7})
```

```python
from numpy import unique
from numpy import where
from sklearn.cluster import AgglomerativeClustering
from matplotlib import pyplot 
from sklearn.metrics import silhouette_score
from sklearn.metrics import calinski_harabasz_score
from sklearn.metrics import davies_bouldin_score

model = AgglomerativeClustering(n_clusters=6)
yhat = model.fit(Z)
yhat_2 = model.fit_predict(Z)
clusters = unique(yhat)

score_AGClustering_s = silhouette_score(Z, yhat.labels_, metric='euclidean')
score_AGClustering_ch = calinski_harabasz_score(Z, yhat.labels_)
score_AGClustering_db = davies_bouldin_score(Z, yhat.labels_)

print('Silhouette Score: ', score_AGClustering_s)
print('Calinski-Harabasz Score: ', score_AGClustering_ch)
print('Davies-Boulding Score: ', score_AGClustering_db)
```
